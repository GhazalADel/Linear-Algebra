```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([5,3])
v2 = np.array([-2,-4])

#we have to define np.array because if we don't `+` here will concate two vectors.
v3 = v1 + v2

#plot
plt.plot([0,v1[0]],[0,v1[1]],'b',label='v1')
plt.plot([0,v2[0]]+v1[0],[0,v2[1]]+v1[1],'r',label='v2')
plt.plot([0,v3[0]],[0,v3[1]],'k',label='v1+v2')

plt.legend() # to show labels
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
plt.show()
```
