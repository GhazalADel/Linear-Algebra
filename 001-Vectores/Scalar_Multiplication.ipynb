```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define vector and scalar
v1 = np.array([2,-1])
l = -.3
v1m = v1*l

#plot
plt.plot([0,v1[0]],[0,v1[1]],'b',label='v1')
plt.plot([0,v1m[0]],[0,v1m[1]],'r:',label='v1m')

plt.axis('square')
axlim=max([abs(max(v1)),abs(max(v1m))])*1.5 # dynamix axis limitation based on vectors
plt.axis((-axlim,axlim,-axlim,axlim))
plt.grid()
plt.show()
```
