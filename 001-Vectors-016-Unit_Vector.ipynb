```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# vector
v1 = np.array([-3,6])
# find scalar
u = 1/(np.linalg.norm(v1))
# unit vector
uv = v1*u
print(np.linalg.norm(uv))

# plot
plt.plot([0,uv[0]],[0,uv[1]],'r--',label='v1-norm',linewidth=5)
plt.plot([0,v1[0]],[0,v1[1]],'b--',label='v1',linewidth=2)

# axis square
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
plt.legend()
plt.show()
```
